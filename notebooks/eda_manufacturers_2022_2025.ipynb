```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/csv/reasons_with_year.csv', parse_dates=['date'])

# Display the first few rows of the DataFrame
summary = pd.read_csv('../data/csv/category_summary_2022_2025.csv')
```

```python
top_mf = (
    df['manufacturer']
    .value_counts()
    .head(10)
    .reset_index()
    .rename(columns={'index': 'manufacturer', 'manufacturer': 'count'})
)
```

```python
top_mf = df['manufacturer'].value_counts().reset_index()
top_mf.columns = ['manufacturer', 'count']
top_mf = top_mf.head(10)

```

```python
top_mf 
```

```python
#plot top 10 manufacturers by number of recalls
sns.set(style='whitegrid')

plt.figure(figsize=(12, 6))
sns.barplot(data=top_mf, x='count', y='manufacturer', palette='viridis')
plt.title('Top 10 Manufacturers by Number of Recalls (20222025)')
plt.xlabel('Number of Recalls')
plt.ylabel('Manufacturer')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```
