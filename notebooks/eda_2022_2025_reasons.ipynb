```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
print(os.getcwd())
```

```python
df = pd.read_csv('../data/csv/reasons_with_year.csv', parse_dates=['date'])

# Display the first few rows of the DataFrame
summary = pd.read_csv('../data/csv/category_summary_2022_2025.csv')
```

```python
df.dropna(subset=['year'])
```

```python
pivot_df = df.pivot_table(
    index='year',
    columns='recall_category',
    fill_value=0,
    aggfunc='size'
)
```

```python
pivot_df.plot(kind='bar',
stacked=True,
colormap='Set2',
figsize=(12, 8),
)
plt.title = 'Drug Recall Reasons by Year(2022-2025)'
plt.xlabel('Year')
plt.ylabel('Number of Recalls')
plt.legend(title='Recall Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
top_categories = df['recall_category'].value_counts().nlargest(5).index.tolist()

top_categories #overall top 5 categories 
```

```python
df_top = df[df['recall_category'].isin(top_categories)]

df_top 
```

```python
#Group by year and category
trend_df = df_top.groupby(['year', 'recall_category']).size().unstack(fill_value=0)
```

```python
#plot trend of top 5 categories

trend_df.plot(marker='o', figsize=(10, 6))
plt.title = 'Trend of Top 5 Drug Recall Categories (2022â€“2025)'
plt.xlabel('Year')
plt.ylabel('Number of Recalls')
plt.legend(title='Recall Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#TRENDS PER QUARTER

df['date'] = pd.to_datetime(df['date'], errors='coerce') #ensures date is in datetime format
df['quarter'] = df['date'].dt.to_period('Q') #define quarter
```

```python
top_5 = df['recall_category'].value_counts().nlargest(5).index.tolist()

top_5
```

```python
df_top5 = df[df['recall_category'].isin(top_5)]
```

```python
quarterly_trends = (
    df_top5.groupby(['quarter', 'recall_category'])
    .size()
    .unstack(fill_value=0)
    .sort_index()
) 
```

```python
#PLOT
quarterly_trends.plot(marker='o', figsize=(12, 6))
plt.title = 'Quarterly Trend of Top 5 Recall Categories (2022â€“2025)'
plt.xlabel('Quarter')
plt.ylabel('Number of Recalls')
plt.grid(True)
plt.legend(title='Recall Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
